In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
import numpy as np

# Load CSV files into DataFrames
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

# Concatenate DataFrames
df1_2 = pd.concat([df1, df2], axis=0).reset_index(drop=True)
df1_2_3 = pd.concat([df1_2, df3], axis=0).reset_index(drop=True)

# Perform gender mapping 1
gender_mapping = {'m': 'M', 'male': 'M', 'f': 'F', 'female': 'F', 'femal': 'F'}
df1_2_3['GENDER'] = df1_2_3['GENDER'].str.strip().str.lower().map(gender_mapping)

# State correction 2 
state_mapping = {'WA': 'Washington', 'Cali': 'California', 'AZ': 'Arizona'}
df1_2_3['ST'] = df1_2_3['ST'].str.strip().map(state_mapping)

# Education correction 3 
edu_mapping = {'Bachelors': 'Bachelor'}
df1_2_3['Education'] = df1_2_3['Education'].str.strip().map(edu_mapping)

# Vehicle class correction 4 
class_mapping = {'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'}
df1_2_3['Vehicle Class'] = df1_2_3['Vehicle Class'].map(class_mapping)

#correcting LTV % 5 
df1_2_3['Customer Lifetime Value'] = df1_2_3['Customer Lifetime Value'].str.replace('%', '')
df1_2_3['Customer Lifetime Value'] = pd.to_numeric(df1_2_3['Customer Lifetime Value'], errors='coerce')

display (df1_2_3)
df1_2_3['Customer Lifetime Value'].dtype
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here
url4 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
mkt_df = pd.read_csv (url4)
mkt_df.rename(columns={'gender':'GENDER', 'customer': 'Customer', 'state':'ST','customer_lifetime_value':'Customer Lifetime Value', 'education':'Education', 'income':'Income', 'monthly_premium_auto':'Monthly Premium Auto', 'number_of_open_complaints':'Number of Open Complaints', 'policy_type':'Policy Type', 'total_claim_amount':'Total Claim Amount','vehicle_class':'Vehicle Class', 'unnamed:_0':'Index', 'employmentstatus': 'Employment Status', 'total_claim_amount': 'Total Claim Amount'}, inplace=True)
mkt_df['Customer Lifetime Value'] = pd.to_numeric(mkt_df['Customer Lifetime Value'], errors='coerce')
mkt_df1_2_3 = pd.merge(df1_2_3,mkt_df, on =['GENDER','Customer','ST','Customer Lifetime Value','Education','Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Total Claim Amount','Vehicle Class'], how='outer')
display (mkt_df1_2_3)
mkt_df1_2_3.columns
mkt_df1_2_3['Total Claim Amount'].dtype 
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

Agent is the leading sales channel in total revenue 

```python
sales_pivot_df = mkt_df1_2_3.pivot_table(index='sales_channel', values='Total Claim Amount', aggfunc='sum')
sales_pivot_df = sales_pivot_df.round(2)
display(sales_pivot_df)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

high school or college eduction level both in male & female have the highest average lifetime value, M: 693 > F: 625 

```python
#mean of gender & education 
ltv_pivot_df = mkt_df1_2_3.pivot_table(index='Customer Lifetime Value', columns=['GENDER', 'Education'], values='Total Claim Amount', aggfunc='sum')
display (ltv_pivot_df) 
ltv_pivot_df.describe (include = 'all')
```
